# 1. Matrix

In this course and a large variety of practical settings of data analysis, data are stored in matrices. Define $A_{n\times m}=(A_{i,j})_{n\times m}$ as a matrix of $n$ rows and $m$ columns, where $n$ and $m$ are positive integers. For example, below we have a 2-by-3 matrix $A$:
$$A=\left(\begin{matrix}
A_{11}&A_{12}&A_{13}\\
A_{21}&A_{22}&A_{23}
\end{matrix}\right)=\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)$$
In practice, the dimension of a matrix can be huge as we are in a big-data array. 

We say two matrices $A=B$ if each entry of $A$ is the same as the corresponding entry of $B$. That is
$$A_{ij}=B_{ij}\mbox{ for all }1\le i\le n,1\le j\le m$$

### Matrix Operations

We describe some commonly used matrix operations. The first one is transpose. We denote the transpose of matrix $A$ as $A^T$. If $B=A^T$, then $B_{ij}=A_{ji}$ for any $i$ and any $j$. Therefore, if $A$ is $n$ by $m$, then $A^T$ is $m$ by $n$. For example, if 
$$A=\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right),\mbox{ then }A^T=\left(\begin{matrix}
2&3\\
0&1\\
-1&2
\end{matrix}\right)$$

The second operation for matrix is summation. Assume $A$ and $B$ are two matrices of the same dimensions ($n\times m$), then the sum of $A$ and $B$ is an $n$ by $m$ matrix satisfying
$$A+B=\left(\begin{matrix}
A_{11}&A_{12}&...&A_{1m}\\
A_{21}&A_{22}&...&A_{2m}\\
...&...&...&...\\
A_{n1}&A_{n2}&...&A_{nm}
\end{matrix}\right)+\left(\begin{matrix}
B_{11}&B_{12}&...&B_{1m}\\
B_{21}&B_{22}&...&B_{2m}\\
...&...&...&...\\
B_{n1}&B_{n2}&...&B_{nm}
\end{matrix}\right)=\left(\begin{matrix}
A_{11}+B_{11}&A_{12}+B_{12}&...&A_{1m}+B_{1m}\\
A_{21}+B_{21}&A_{12}+B_{22}&...&A_{1m}+B_{2m}\\
...&...&...&...\\
A_{n1}+B_{n1}&A_{n2}+B_{n2}&...&A_{nm}+B_{nm}
\end{matrix}\right)$$
For example,
$$\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)+\left(\begin{matrix}
-1&3&1\\
2&-2&4
\end{matrix}\right)=\left(\begin{matrix}
1&3&0\\
5&-1&6
\end{matrix}\right)$$
The matrix substraction operation can be defined as:
$$A-B=A+(-B)$$
where the dimensions of $A$ and $B$ are the same.

Similarly, we can define the scalar multiplication of a matrix, which is a matrix of the same dimension: if $$A=\left(\begin{matrix}
A_{11}&A_{12}&...&A_{1m}\\
A_{21}&A_{22}&...&A_{2m}\\
...&...&...&...\\
A_{n1}&A_{n2}&...&A_{nm}
\end{matrix}\right)$$ 
then
$$kA=\left(\begin{matrix}
kA_{11}&kA_{12}&...&kA_{1m}\\
kA_{21}&kA_{22}&...&kA_{2m}\\
...&...&...&...\\
kA_{n1}&kA_{n2}&...&kA_{nm}
\end{matrix}\right)$$
For example, 
$$A=2\times\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)=\left(\begin{matrix}
4&0&-2\\
6&2&4
\end{matrix}\right)$$



Next, we consider the multiplication of matrices. Consider two matrices $A$ and $B$, where $A$ is $p$ by $q$ and $B$ is $q$ by $r$. Then, $M=AB$ is a $p$ by $r$ matrix defined as:
$$M_{ij}=\sum_{k=1}^qA_{ik}B_{kj}\mbox{ for all }i\mbox{ and }j$$
For example, if $A=\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)$ and $B=\left(\begin{matrix}
1&-4\\
2&-1\\
3&-2
\end{matrix}\right)$, then
$$AB=\left(\begin{matrix}
2\times1+0\times2+(-1)\times3&2\times(-4)+0\times(-1)+(-1)\times(-2)\\
3\times1+1\times2+2\times3&3\times(-4)+1\times(-1)+2\times(-2)
\end{matrix}\right)=\left(\begin{matrix}
-1&-6\\
11&-17
\end{matrix}\right)$$

One should note that, in general $AB\ne BA$. In fact, their dimensions may not even match. For example, still assume $A=\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)$ and $B=\left(\begin{matrix}
1&-4\\
2&-1\\
3&-2
\end{matrix}\right)$. Then, $AB=\left(\begin{matrix}
-1&-6\\
11&-17
\end{matrix}\right)$ and $BA=\left(\begin{matrix}
-10&-4&-9\\
1&-1&-4\\
0&-2&-7
\end{matrix}\right)$.


Matrices enable us to streamline the representations of functions, ineuqlities and equalities. For example:

- $f(x)=\sum_{i=1}^nc_ix_i$. Then, we can write $c=(c_1,c_2,...,c_n)^T$ and $x=(x_1,x_2,...,x_n)^T$. We have $f(x)=c^Tx$. For instance, if $f(x)=2x_1+3x_2$, then $f(x)=(2,3)\left(\begin{matrix}
x_1\\
x_2
\end{matrix}\right)$
- $\sum_{j=1}^nA_{ij}x_j\le B_i$ for $i=1,2,...,m$. We can write $A=(A_{ij})_{m\times n}$, $x=(x_1,x_2,...,x_n)^T$, and $B=(B_1,B_2,...,B_m)$. We have $Ax\le B$. For instance, $2x_1+x_2\le 3$ and $3x_1-2x_2\le 1$ can be written in matrix form as $\left(\begin{matrix}
2&1\\
3&-2
\end{matrix}
\right)\left(\begin{matrix}
x_1\\
x_2
\end{matrix}\right)\le\left(\begin{matrix}
3\\
1
\end{matrix}\right)$.
- $\sum_{j=1}^na_{ij}x_j\le b_i$ for $i=1,2,...,m$. We can write $a=(A_{ij})_{m\times n}$, $x=(x_1,x_2,...,x_n)^T$, and $b=(b_1,b_2,...,b_m)$. We have $ax= B$. For instance, $x_1-2x_2= 1$ and $3x_1+2x_2= 4$ can be written in matrix form as $\left(\begin{matrix}
1&-2\\
3&2
\end{matrix}
\right)\left(\begin{matrix}
x_1\\
x_2
\end{matrix}\right)=\left(\begin{matrix}
1\\
4
\end{matrix}\right)$.

Throughout this course, we will work extensively with matrices to represent and analyze data. In Python, matrices are stored in the format of arrays, using the NumPy and Pandas libraries.

# 3. Data Analysis with NumPy

NumPy is a python library for working with arrays. It also has functions convenient for linear algebra, fourier transformation, and matrics operations. In Python, we actually have the data type for arrays (called lists), but they are slow to process. NumPy aims to provide an array object that is up to 50x faster than the traditional Python lists. The array object in NumPy is called ndarray. It provides a lot of supporting functions that facilitates working with ndarray. This section only introduces the very basics. You could refer to numerous online NumPy tutorials to learn more. 

To begin with, we first create 1- and 2-dimensional arrays. The method in NumPy `array()` creates an ndarray object, whereas the method `shape` returns the  lengths of the corresponding array dimensions.

````python
## Create a vector:
vec = np.array([1, 2, 3, 4, 5])

print(vec)
print('The shape of the vector is',vec.shape)

## Create a matrix:

mat = np.array([[1,2,3],[4,5,6]])
print('\n')
print(mat)
print('The shape of the matrix is', mat.shape)
````

Next, we try to access different entries of an array.

````python
# For a vector, the index starts with 0. Index -1 means the last element.

print(vec[0],vec[2],vec[-1])

# For a matrix, the index specifies both the row and the column.

print(mat[0,2])
print(mat[1,:])
print(mat[:,-1])
````

Next, we reshape an array using the method `reshape()`.

````python
# Reshape mat into a vector and into a 1 by 6, and a 3 by 2 matrices.
mat1 = mat.reshape(1,6)
mat2 = mat.reshape(3,2)
print(mat1)
print(mat2)
````

Next, we consider joining multiple arrays into a single one. In NumPy, we use the `concatenate()` function to join the arrays. We also should specify along which axis we want to join the arrays. If axis is not explicitly passed, its default value is 0.

````python
# Join arrays:

arr1 = np.array([[1, 2, 3]])

arr2 = np.array([[4, 5, 6]])

arr_join0 = np.concatenate((arr1, arr2), axis=0)

arr_join1 = np.concatenate((arr1, arr2), axis=1)
print('The array joined along axis 0:','\n',arr_join0)
print('Shape of the array joined along axis 0:',arr_join0.shape)
print('\n')
print('The array joined along axis 1:',arr_join1)
print('Shape of the array joined along axis 1:',arr_join1.shape)
````

Next, we discuss how to join DataFrames using indices. We will use the `merge()` function in Pandas. We want to highlight that joining is arguably the most important operation in data analysis since it combines the information from different data sets in an integral fashion.

````python
## Create 2 DataFrames
df_BA = pd.DataFrame({'Name':['Adam','Bob','Charles','Danniel','Edward','Frank'],\
                      'BA_Grade':['B-','A','A-','B+','B','A']})
df_Finance = pd.DataFrame({'Name':['Amy','Bob','Catherine','Danniel','Edward','Fiona'],\
                      'Finance_Grade':['A','B+','B-','A','B','B-']})
df_BA
````

````python
df_Finance
````

````python
#Outer join

df_outer = pd.merge(df_BA,df_Finance,on = ['Name'],how = 'outer')
df_outer
````

````python
# Left join
df_left = pd.merge(df_BA,df_Finance,on = ['Name'],how = 'left')
df_left
````

````python
# Right join
df_right = pd.merge(df_BA,df_Finance,on = ['Name'],how = 'right')
df_right
````

````python
# Inner join
df_inner = pd.merge(df_BA,df_Finance,on = ['Name'],how = 'inner')
df_inner
````

The next operation of Pandas we want to discuss in this document is the aggregate data using the `group_by()` function. To illustrate this application, we create a DataFrame of T-shirt sales data first. The function `MultiIndex.from_arrays()` creates multiple indices from an array.

````python
# Create a dictionary.
data_dict = {'color' : ['black', 'white', 'black', 'white', 'black',\
                      'white', 'black', 'white', 'black', 'white'],\
            'size' : ['S','M','L','M','L','S','S','XL','XL','M'],\
            'price': [99,79,108,89,99,100,105,111,86,94],\
            'quantity': [1210,1340,2410,1458,1124,2217,989,1009,1342,1077]}
array=[['A','B','B','B','C','A','B','A','C','C'],['JP','CN','US','US','US','CN','CN','CA','JP','CA']]

# Create the two indices.
index = pd.MultiIndex.from_arrays(array, names=['class', 'country'])

# Create the DataFrame.
data_df = pd.DataFrame(data_dict,index=index)
data_df
````

````python
## We group the data by size

df_size = data_df.groupby('size')

# Calculate the average price and quantity of each size

df_size.mean().add_prefix('average_')
````

We now summarize other aggregation functions that could be used for the `group_by()` function.

- `count`: Count the number of elements in the group that are not NA.
- `mean`: Compute the average value of the group.
- `median`: Compute the median value of the group.
- `std`: Compute the standard deviation of the group.
- `var`: Compute the variance of the group.
- `min`: Compute the minimum value of the group.
- `max`: Compute the maximum value of the group.
- `prod`: Compute the product of all values in the group.
- `first`: Find the first value in the group that is not NA.
- `last`: Find the last value in the group that is not NA.

Finally, the function to store data into a csv file is `to_csv()`.

Then, the bar plot.

````python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N) # the locations for the groups. np.range(N) returns an array [0,1,...,N-1]
width = 0.35 # The width of the bar in the bar plot.
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, menMeans, width, color='r')
ax.bar(ind, womenMeans, width,bottom=menMeans, color='b') #bottom specifies where the bottom of the bar is.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(['G1', 'G2', 'G3', 'G4', 'G5'])
ax.set_yticks(np.arange(0, 81, 10))  
#np.arange(0,81,10) returns an array starting at 0, stopping at 81 (excluded), with a gap of 10
ax.legend(labels=['Men', 'Women'])
plt.show()
````
